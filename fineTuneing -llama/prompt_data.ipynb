```python
import os
import sys
from pathlib import Path

# Ensure local `pricer/` package is importable regardless of notebook cwd
for _p in [
    Path.cwd(),
    Path.cwd().parent,
    Path.cwd() / "code",
    Path.cwd().parent / "code",
]:
    if (_p / "pricer").is_dir() and str(_p) not in sys.path:
        sys.path.insert(0, str(_p))
        break

from dotenv import load_dotenv
from huggingface_hub import login
from pricer.items import Item
from tqdm.notebook import tqdm
from transformers import AutoTokenizer
import matplotlib.pyplot as plt
```

```python
hf_token = os.getenv("HF_TOKEN") or os.getenv("HUGGINGFACEHUB_API_TOKEN")
if not hf_token:
    raise RuntimeError(
        "Missing Hugging Face token. Set HF_TOKEN (or HUGGINGFACEHUB_API_TOKEN) in code/.env"
    )

login(hf_token, add_to_git_credential=True)
```

```python
username = "ed-donner"
dataset = f"{username}/items_lite" 
train, val, test = Item.from_hub(dataset)
items = train + val + test
```

```python
print(f"Loaded {len(train):,} training items, {len(val):,} validation items, {len(test):,} test items")
```

```python
BASE_MODEL = "meta-llama/Llama-3.2-3B"

tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
```

```python
token_counts = [item.count_tokens(tokenizer) for item in tqdm(items)]
```
